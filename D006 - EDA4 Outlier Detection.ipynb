## Summary of D6 Outlier Detection

### Outliers
An outlier is an observation that is unlike the other observations.
It is rare, or distinct, or does not fit in some way.

Outliers can have many causes, such as:
 - Measurement or input error.
 - Data corruption.
 - True outlier observation
 
### Ways to detect outliers
1. Plotting: using box plot or scatter plot
2. Observer Min, Max, Mean, Median and std
3. Guassian-like distribution can use Z-Score (within 3 std)
4. IQR (1.5) can apply on any distribution

### Handle with outliers
1. Add one more column to reflect outliers
2. Replace outliers with median, min, or any meaningful values

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
cdf = app_train['AMT_INCOME_TOTAL'].value_counts()
cdf = cdf.sort_index().cumsum()*1./len(cdf)

plt.plot(list(cdf.index), cdf/cdf.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf.index)), cdf/cdf.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
